```python
economic_data.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
# Adding new columns for each war
economic_data['Korean War'] = False
economic_data['Vietnam War'] = False
economic_data['Persian Gulf War'] = False
economic_data['War in Afghanistan'] = False
economic_data['Iraq War'] = False
economic_data['Russia Ukraine War'] = False

# Defining the time ranges for each war
war_periods = {
    'Korean War': ('1950-06-25', '1953-07-27'),
    'Vietnam War': ('1965-03-08', '1975-04-30'),
    'Persian Gulf War': ('1990-08-02', '1991-02-28'),
    'War in Afghanistan': ('2001-10-07', '2021-08-30'),
    'Iraq War': ('2003-03-20', '2011-12-18'),
    'Russia Ukraine War': ('2022-02-24', '2023-12-31')  # End date is set to my last update
}

# Converting the 'Quarter' column to datetime for comparison
economic_data['Quarter'] = pd.to_datetime(economic_data['Quarter'])

# Marking 'True' or 'False' for each war based on the date
for war, (start_date, end_date) in war_periods.items():
    economic_data[war] =  economic_data['Quarter'].between(start_date, end_date)

economic_data.head()
```

## Based on historical Consumer Price Index and unemployment rate data, predict future inflation trends

```python
from sklearn.metrics import mean_squared_error
import numpy as np

# Isolating the 'Inflation Rate' column
inflation_data = economic_data['Inflation Rate'].dropna().values

# Creating the naive forecast
# The forecast for each point is simply the previous observed value
naive_forecast = inflation_data[:-1]

# Actual values to compare against (excluding the first value as there's no prediction for it)
actual_values = inflation_data[1:]

# Calculate the Mean Squared Error (MSE)
mse = mean_squared_error(actual_values, naive_forecast)

# Calculate the Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)

rmse
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam


# Load your data
# df = pd.read_csv('your_data.csv')  # Replace with your actual data file

# For the sake of example, I'll use a simulated small dataset
data = {
    "Consumer Price Index": [164.6, 166.2, 168.0, 170.5, 172.3, 174.1],
    "UNRATE": [4.3, 4.26, 4.2, 4.1, 4.05, 4.0],
    "Inflation Rate": [1.66, 2.13, 2.50, 2.85, 3.10, 3.30]
}
df = pd.DataFrame(data)

# Preprocessing
X = df[["Consumer Price Index", "UNRATE"]]
y = df["Inflation Rate"]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Model Creation
model = Sequential()
model.add(Dense(64, activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='linear'))

# Compile the model
model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')

# Model Training
model.fit(X_train, y_train, epochs=100, batch_size=10, verbose=1)

# Model Evaluation
loss = model.evaluate(X_test, y_test, verbose=1)
print(f'Test Loss: {loss}')


```
